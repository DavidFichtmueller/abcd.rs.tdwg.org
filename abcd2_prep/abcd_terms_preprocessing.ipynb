```python
terms = terms.rename(columns={"type": "xsd_data_type",
                             "group": "group_old",
                             "subgroup": "subgroup_old"})

terms = terms.rename(columns={"identifier": "tdwgutil_xpath", 
                      "short_name": "term_localName", 
                      "replaces": "deprecated_xpath", 
                      "parent_short_name": "parent", 
                      "element_name": "label", 
                      "Export_documentation": "definition", 
                      "usage_notes": "usage", 
                      #"": "notes", 
                      "Export_example": "examples",
                      "group_new": "tdwgutility_organizedInClass", 
                      "subgroup_new": "subgroup", 
                      "tdwg_export_type": "type", 
                      "required": "tdwgutility_required", 
                      "tdwgutil_repeatable": "tdwgutil_repeatable", 
                      "date_created": "term_created", 
                      "date_modified": "term_modified"})

terms
```

```python
for index,term in terms.iterrows():
    if term["subgroup"] != "":
        term["tdwgutility_organizedInClass"] = term["tdwgutility_organizedInClass"] + "-" + term["subgroup"]
        
terms[terms["subgroup"] != ""][["tdwgutility_organizedInClass","subgroup"]]
```
